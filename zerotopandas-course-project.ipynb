```python
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    # Iterate over the column
    for idx, value  in col_series[col_series.notnull()].items():
        # Break each value into list of options
        for option in value.split(';'):
            # Add the option as a column to result
            if not option in result_df.columns:
                options.append(option)
                result_df[option] = False
            # Mark the value in the option column as True
            result_df.at[idx, option] = True
    return result_df[options]
```

# Streams

# Q3. Is there a difference in streams choosed by students?

```python
stream_type_df = split_multicolumn(survey_df.hsc_s)
```

```python
stream_type_df
```

```python
stream_type_totals = stream_type_df.sum().sort_values(ascending=False)
stream_type_totals
```

```python
stream_type2_df = split_multicolumn(survey_df.degree_t)
```

```python
stream_type2_df
```

```python
stream_type2_totals = stream_type2_df.sum().sort_values(ascending=False)
stream_type2_totals
```

```python
fig, axis = plt.subplots(1, 2, figsize=(24, 12))
axis[0].set_title("stream mostly choosed by student in hsc",fontdict={'fontsize': 25})
sns.barplot(x=stream_type_totals,y=stream_type_totals.index, ax=axis[0])
axis[0].set_ylabel("stubjects in hsc",fontsize=25)
axis[0].set_xlabel("number of students",fontsize=25)
axis[1].set_title("stream choosen by students in degree",fontdict={'fontsize': 25})
sns.barplot(x=stream_type2_totals, y=stream_type2_totals.index)
axis[1].set_ylabel("subjects in degree",fontsize=25)
axis[1].set_xlabel("number of students",fontsize=25);
```

<p style="font-size:20px">From this graph, we are getting to the point that number of students who have opted science in their higher secondary studies are more whereas when they go to college they sifted to other degrees or it might be a case that those students actually didn't get admitted in college, therefore, the sciandtech degree have less students. 
Same goes for the commerce and commandmgmt degree, the case is just reverse of science.</p>
